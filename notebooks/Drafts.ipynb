```python
def daily_aggregation(x):
    names = {
        'temp_daily_min': x['temp_min'].min()-273.15,
        'temp_daily_max':  x['temp_max'].max()-273.15,
        'pressure_min' : x['pressure'].min(),
        'pressure_max' : x['pressure'].max(),
        'clouds_min' : x['clouds_all'].min(),
        'clouds_max' : x['clouds_all'].max()
    }

    return pd.Series(names, index=reversed(sorted(names.keys())))

daily_weather.apply(daily_aggregation)
```

```python
x_column = 'temp_daily_min'
y_column = 'Pulse'

plt.gcf().set_size_inches(10,10)
legend_list = list()
for i in exploration_patient_ids:
    cur_ds = patients_and_weather_df[patients_and_weather_df['Merida ID'] == i]
    plt.scatter(cur_ds[x_column], cur_ds[y_column])
    non_nan_selector = ~np.isnan(cur_ds[x_column]) & ~np.isnan(cur_ds[y_column])
    non_nan_x = cur_ds[x_column][non_nan_selector]
    non_nan_y = cur_ds[y_column][non_nan_selector]
    cor_coef = np.corrcoef(non_nan_x, non_nan_y)[0][1]
    legend_list.append(str(cor_coef))
plt.legend(legend_list)
plt.show()
```
